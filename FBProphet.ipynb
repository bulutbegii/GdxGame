```python
from prophet import Prophet
import yfinance as yf
```

```python
df=yf.download('BTC-USD',"2015-01-01","2024-07-24")
```

```python
df=df[['Close']]
```

```python
df=df.reset_index()
```

```python
df.columns=['ds','y']
```

```python
df.head(3)
```

```python
model=Prophet()
```

```python
model.fit(df)
```

```python
gelecek=model.make_future_dataframe(periods=10)
```

```python
tahmin=model.predict(gelecek)
```

```python
model.plot(tahmin);
```

```python
model.plot_components(tahmin);
```
